```python
import matplotlib.pyplot as plt
import pickle
timelist = [i * 0.5 for i in range(180)]
# Open the pickle file in binary mode for reading
with open('Heartrate_Wu.pkl', 'rb') as file:
    # Load the contents of the pickle file
    data1 = pickle.load(file)
plt.plot(timelist,data1,color='blue',label='Wu')
plt.xlabel('Time (seconds)')  # 更新横轴标签为秒
plt.ylabel('Heart Rate')
plt.title('Heart Rate Variation(seconds)')

# Open the pickle file in binary mode for reading
with open('Heartrate_Wang.pkl', 'rb') as file:
    # Load the contents of the pickle file
    data2 = pickle.load(file)
plt.plot(timelist,data2,color='red',label='Wang')
plt.xlabel('Time (seconds)')  # 更新横轴标签为秒
plt.ylabel('Heart Rate')
plt.title('Heart Rate Variation(seconds)')



# Open the pickle file in binary mode for reading
with open('Heartrate_Kang.pkl', 'rb') as file:
    # Load the contents of the pickle file
    data3 = pickle.load(file)
plt.plot(timelist,data3,color='y',label='Kang')
plt.xlabel('Time (seconds)')  # 更新横轴标签为秒
plt.ylabel('Heart Rate')
plt.title('Heart Rate Variation(seconds)')



# Open the pickle file in binary mode for reading
with open('Heartrate_Yang.pkl', 'rb') as file:
    # Load the contents of the pickle file
    data4 = pickle.load(file)
plt.plot(timelist,data4,color='k',label='Yang')
plt.xlabel('Time (seconds)')  # 更新横轴标签为秒
plt.ylabel('Heart Rate')
plt.title('Heart Rate Variation(seconds)')

plt.legend()
plt.show()
```

```python
print('Heart rate for Wu =',sum(data1) / len(data1),'BPM')
print('Heart rate for Wang =',sum(data2) / len(data2),'BPM')
print('Heart rate for Kang =',sum(data3) / len(data3),'BPM')
print('Heart rate for Yang =',sum(data4) / len(data4),'BPM')
```

```python
print('Overall heart rate=',(sum(data1)+sum(data2)+sum(data3)+sum(data4))/(4*len(data1)),'BPM')
```
